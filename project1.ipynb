```python
import hid
import time

VENDOR_ID = 0x3633
PRODUCT_ID = 0x0002

dev = hid.device()
dev.open(VENDOR_ID, PRODUCT_ID)
print("Connected to:", dev.get_product_string())

last_trimmed = None

try:
    while True:
        response = dev.read(64)
        trimmed = response
        while trimmed and trimmed[-1] == 0:
            trimmed = trimmed[:-1]
        if trimmed != last_trimmed:
            print("Reading:", trimmed)
            last_trimmed = trimmed
        time.sleep(0.5)
finally:
    dev.close()
```

```python
import hid
import time

VENDOR_ID = 0x3633
PRODUCT_ID = 0x0002

dev = hid.device()
dev.open(VENDOR_ID, PRODUCT_ID)
print("Connected to:", dev.get_product_string())

last_trimmed = None

def get_mode(reading):
    if len(reading) >= 2:
        if reading[0] == 16 and reading[1] == 76:
            return "Utilization"
        elif reading[0] == 16 and reading[1] == 19:
            return "Temperature (C)"
        elif reading[0] == 16 and reading[1] == 35:
            return "Temperature (F)"
    return "Unknown"

try:
    while True:
        response = dev.read(64)
        trimmed = response
        while trimmed and trimmed[-1] == 0:
            trimmed = trimmed[:-1]
        if trimmed != last_trimmed:
            mode = get_mode(trimmed)
            value = trimmed[-1] if trimmed else None
            print(f"Mode: {mode} | Reading: {trimmed} | Value: {value}")
            last_trimmed = trimmed
        time.sleep(0.5)
finally:
    dev.close()
```

```python
import hid
import time
import psutil

VENDOR_ID = 0x3633
PRODUCT_ID = 0x0002

dev = hid.device()
dev.open(VENDOR_ID, PRODUCT_ID)
print("Connected to:", dev.get_product_string())

try:
    while True:
        cpu_percent = int(psutil.cpu_percent())
        # Build output report for utilization
        out_report = [16, 76, 1, 0, 0, cpu_percent] + [0] * 58
        dev.write(out_report)
        print(f"Sent Utilization: {cpu_percent}%")
        time.sleep(1)
finally:
    dev.close()
```

```python
import hid
import time
import wmi

VENDOR_ID = 0x3633
PRODUCT_ID = 0x0002

def get_cpu_temp():
    w = wmi.WMI(namespace="root\\OpenHardwareMonitor")
    sensors = w.Sensor()
    for sensor in sensors:
        if sensor.SensorType == u'Temperature' and 'CPU' in sensor.Name:
            return int(sensor.Value)
    return 0

dev = hid.device()
dev.open(VENDOR_ID, PRODUCT_ID)
print("Connected to:", dev.get_product_string())

try:
    while True:
        temp_c = get_cpu_temp()
        tens = temp_c // 10
        ones = temp_c % 10
        # Try putting tens and ones in positions 4 and 5
        out_report = [16, 19, 1, 0, tens, ones] + [0] * 58
        dev.write(out_report)
        print(f"Sent Temperature (C): {temp_c}°C (tens={tens}, ones={ones})")
        time.sleep(1)
finally:
    dev.close()
```

```python
import hid
import time
import wmi

VENDOR_ID = 0x3633
PRODUCT_ID = 0x0002

def get_cpu_temp():
    w = wmi.WMI(namespace="root\\OpenHardwareMonitor")
    sensors = w.Sensor()
    for sensor in sensors:
        if sensor.SensorType == u'Temperature' and 'CPU' in sensor.Name:
            return int(sensor.Value)
    return 0

dev = hid.device()
dev.open(VENDOR_ID, PRODUCT_ID)
print("Connected to:", dev.get_product_string())

try:
    while True:
        temp_c = get_cpu_temp()
        temp_c = min(max(temp_c, 0), 99)  # Clamp to 0–99°C
        tens = temp_c // 10
        ones = temp_c % 10
        out_report = [16, 19, 1, 0, tens, ones] + [0] * 58
        dev.write(out_report)
        print(f"Sent Temperature (C): {temp_c}°C (tens={tens}, ones={ones})")

        # Read back what is shown
        response = dev.read(64)
        trimmed = response
        while trimmed and trimmed[-1] == 0:
            trimmed = trimmed[:-1]
        print(f"Received: {trimmed}")

        time.sleep(2)
finally:
    dev.close()
```
